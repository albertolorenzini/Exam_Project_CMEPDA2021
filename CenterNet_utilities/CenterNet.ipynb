#5. Training
These three blocks include the instructions to start the training.

IMPORTANT: you have to run only the correct one, depending on the pre-treined chosen weights.
If "out of memory" error occurs, try to change "batch_size" with a lower value.

If Google Colaboratory stops your run before the training has finished, you can restart the training using "--resume" flag. CenterNet will find the latest model with the same exp_id. In order to do it make sure that your Googe Drive disk is still mounted.

coco_dla_2x

```
%cd /home/CenterNet/src
!python main.py ctdet \
  --exp_id coco_dla_2x \
  --batch_size 64 \
  --master_batch 9 \
  --lr 5e-4 \
  --gpus 0 \
  --num_workers 16
```

coco_resdcn101
