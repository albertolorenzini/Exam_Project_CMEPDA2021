#5. Training
These three blocks include the instructions to start the training.

IMPORTANT: you have to run only the correct one, depending on the pre-treined chosen weights.

if error "Out of memory" occurs, then in .cfg-file you should increase "subdivisions", or decrease "batch_size".

If Google Colaboratory stops your run before the training has finished, you have to restart the training starting from the last weights. In order to do that, you have to replace the last command line flag (e.g. "./yolov4.weights") with "../../content/drive/MyDrive/darknet_backup_weights/name_of_last_model.weights". Your last model are saved inside you Drive, so you have to have your Google Drive disk still mounted.

Yolo Pre-Treined obj

```
%cd /home/darknet/
! ./darknet detector train ./data/obj.data ./cfg/yolo-obj.cfg ./yolov4.conv.137 -map -dont_show
```

YoloV4 pre-trained

```
%cd /home/darknet/
! ./darknet detector train ./data/obj.data ./cfg/yolov4-obj.cfg ./yolov4.weights -map -dont_show -clear
```

Tiny-YOLO pre-trained

```
%cd /home/darknet/
! ./darknet detector train ./data/obj.data ./cfg/yolov4-tiny-obj.cfg ./yolov4-tiny.weights -map -dont_show
```
